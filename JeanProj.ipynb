```
import pandas as pd

import matplotlib.pyplot as plt

tabela = pd.read_excel('Dados1.xlsx')

df_plot = tabela[['Cidade', 'População total', 'PIB per capita']]

plt.figure(figsize=(12, 6))

bar_width = 0.4
x = range(len(df_plot))

plt.bar(x, df_plot['População total'], width=bar_width, label='População total', color='skyblue')

plt.bar([i + bar_width for i in x], df_plot['PIB per capita'], width=bar_width, label='PIB per capita', color='orange')

plt.xlabel('Cidade')
plt.ylabel('Valores')
plt.title('População Total e PIB per capita por Cidade')
plt.xticks([i + bar_width / 2 for i in x], df_plot['Cidade'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()

display(tabela)


```
